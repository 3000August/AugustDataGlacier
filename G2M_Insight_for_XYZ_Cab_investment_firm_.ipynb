**COMBINING THE DATASETS**

```python
#the cab_data and transaction_id datasets are related on the 'Transaction ID' column, therefore they can be joined
```

```python
#joining the ab_data and transaction_id datasets
cab_d = cab_data.set_index(['Transaction ID'])
transaction_d = transaction_id.set_index(['Transaction ID'])
cab_trans = cab_d.join(transaction_d)
cab_trans
```

```python
#resetting the index
cab_trans.reset_index(level =['Transaction ID'], inplace = True)
cab_trans
```

```python
#the cab_trans and customer_id datasets are related on the 'Customer ID' column therefore they can be joined
```

```python
#joining the cab_trans and customer_id datasets
joined = cab_trans.set_index([cab_trans.index, 'Customer ID'])
cust_id = customer_id.set_index(['Customer ID'])
data_join = joined.join(cust_id)
data_join
```

```python
#resetting the index
data_join.reset_index(level =['Customer ID'], inplace = True)
data_join
```

```python
#to create the master dataset for exploratory data analysis, we merge the city and data_join dataset as they are related on the 'City' column
```

```python
#creating the master dataset
master_data = pd.merge(data_join, city, on = 'City', how = 'inner')
master_data
```

```python
master_data[['Population', 'Users']].info()
```

```python
#before beginning exploratory data analysis properly, the datatypes of the rows have to be uniform
```

```python
#renaming the columns
for col in master_data.columns:
    if ' ' in col:
        master_data = master_data.rename(columns={col:col.replace(' ','_')})

        master_data
```

```python
#transforming the datatypes in the 'Population' and 'Users' column from objects to integers

master_data['Population'] = master_data['Population'].str.replace(',', '').astype(int)
master_data['Users'] = master_data['Users'].str.replace(',', '').astype(int)
master_data
```

```python
#transforming the 'Date_of_Travel' rows to datetime objects


master_data['Date_of_Travel'] = master_data['Date_of_Travel'].apply(lambda x: datetime.datetime.fromtimestamp((x - 25569) * 86400.0))
master_data
```

# Profit Analysis

```python
#getting the profit on each transction
master_data['Profit'] = master_data['Price_Charged'] - master_data['Cost_of_Trip']
master_data
```

```python
#Total profit of both companies in the given timeframe
tot_prof = master_data.groupby('Company').Profit.sum()
tot_prof
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x=tot_prof.index, y=tot_prof.values)
plt.legend()
plt.title('Total Profits by both companies')
plt.show()
```

```python
#annual profits of both companies
months=[]
years =[]
for i in range(len(master_data['Date_of_Travel'])):
    months.append(master_data['Date_of_Travel'][i].month)
    years.append(master_data['Date_of_Travel'][i].year)
master_data['Month'] = months
master_data['Year'] = years
master_data['Company'] = cab_data['Company']

annual_prof = master_data.groupby(['Year','Company']).Profit.count()

fig, ax = plt.subplots(figsize=(15,5))
annual_prof.unstack().plot(ax=ax, kind = 'bar', color = ['pink','yellow'], title='Annual Profits of Both Companies')
```

```python
#getting the profit per kilometer of both companies
master_data['Profit_Per_KM'] = master_data['Profit']/ master_data['KM_Travelled']

fig, ax = plt.subplots(figsize=(15,6))
fig.suptitle('Profit Per Kilometer Travelled by Both Companies')
sns.histplot(ax=ax, data=master_data , x='Profit_Per_KM' , hue="Company")
plt.show()
```

# *One*

```python
# h0: there is no statistical difference in the profits due to gender
#h1: there is a statistical difference in the profits due to the gender


a = master_data[(master_data.Gender=='Female')&(master_data.Company=='Yellow Cab')].groupby('Transaction_ID').Profit.mean()
b = master_data[(master_data.Gender=='Male')&(master_data.Company=='Yellow Cab')].groupby('Transaction_ID').Profit.mean()
print(a.shape[0],b.shape[0])

from scipy import stats
_, p_value = stats.ttest_ind(a.values,b=b.values,equal_var=True)
if(p_value < 0.05):
    print('We accept alternate hypothesis that there is a statistical difference ')
else:
    print('We accept null hypothesis that there is no statistical difference')
    
print('P value is ', p_value)
```

# Two

```python
# h0: KMs Travelled and Profit gained are not related. (p = 0)
# h1: KMs Travelled and Profit gained are related. (p != 0)

from scipy.stats import pearsonr

stat, p = pearsonr(master_data['KM_Travelled'], master_data['Profit'])
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print('KMs Travelled and Profit gained are not related.')
else:
    print('KMs Travelled and Profit gained are related.')
```
